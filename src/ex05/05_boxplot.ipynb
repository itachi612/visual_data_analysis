<h2>Exercise 05: Boxplot</h2>

```python
import pandas as pd 
import sqlite3
import matplotlib.pyplot as plt
```

### Preprocessing

```python
df = pd.read_csv("../data/ab-test.csv")
df
```

```python
before = [df[(df['time'] == 'before') & (df['group'] == 'control')]['diff'],
        df[(df['time'] == 'before') & (df['group'] == 'test')]['diff']]
after = [df[(df['time'] == 'after') & (df['group'] == 'control')]['diff'],
         df[(df['time'] == 'after') & (df['group'] == 'test')]['diff']]
```

### Result

```python
box_style = dict(
    tick_labels=['control', 'test'],
    patch_artist=True,
    showmeans=True,                                     
    meanprops=dict(marker='^', markerfacecolor='green', markeredgecolor='green'),
    boxprops=dict(facecolor='white', color='green', linewidth=3),
    whiskerprops=dict(color='red', linewidth=2),
    capprops=dict(color='red', linewidth=2),
    medianprops=dict(color='red', linewidth=2),
    flierprops=dict(marker='o', markerfacecolor='none', markeredgecolor='green') # Точки выбросов тоже зеленые
)
```

```python


fig, ax = plt.subplots(1, 2, figsize=(15, 8), sharey=True)

ax[0].boxplot(after, **box_style)
ax[0].set_title("after")

ax[1].boxplot(before, **box_style)
ax[1].set_title("before")

fig.suptitle('Lag before labs deadlines. After and before the Newsfeed', fontsize=15)
ax[0].grid(axis='y', linestyle='-', alpha=0.5)
ax[1].grid(axis='y', linestyle='-', alpha=0.5)
ax[0].set_xlabel('[group]')
ax[1].set_xlabel('[group]')

plt.show()
```
